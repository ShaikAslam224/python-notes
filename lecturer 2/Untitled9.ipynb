```python
import random
name=input("Enter your name: ")
print(f"\nHello {name} welcome to guessing game")
chances=10
number=str(random.randint(0,100))
guessing=""

play="yes"
def playagain():
    global play
    play=input("Do you want to play this game again? (yes/no) :")
    if play =="yes":
        global chances,number,guessing
        chances=10
        number=str(random.randint(0,100))
        guessing=""
    else:
        print("Thank you for playing")
while play=='yes':    
    while chances>0:
        won=True
        for ch in number:
            if ch in guessing:
                print(ch,end="")
            else:
                print('_',end="")
                won=False
                
        if won:
            print("\nyou won the game")
            print(f"\nyour score is {chances*10} ")
            playagain()
            break
        guess=input("\nGuess a two digit number : ")
        guessing+=guess
        if guess < number:
            chances-=1
            print("\nWrong Answer!")
            print(f"\nyou score {chances} chances left")
            print("\nnumber is greater than your number!")
            
            if chances==0:
                print("\nyou lose the game!")
                playagain()
                break
        else:
            chances-=1
            print("\nWrong Answer!")
            print("\nnumber is less than your number!")
            print(f"\nyou have {chances} chances left")
            if chances==0:
                print("\nyou lose the game!")
                playagain()
                break
                

```
