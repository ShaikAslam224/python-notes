```python
name=input("Enter you name: ")
print(f"Hello {name}! welcome to guessing game")
chances=10
word="apple"
guesses="al"   # here guessed words are stored 

# this whole game will run up to 0 chances left
while chances>0:
    won=True
    for ch in word:
        if ch in guesses:
            print(ch,end=" ")
        else:
            print("_",end=" ")
            won=False
    if won:    # if you win this will run
        print("\nYou won the game!!")
        print(f"\nYou score is : {chances*10}")
        break
    # taking a guess word from the user        
    guess=input("\nGuess a character:")
    guesses+=guess
    if guess not in  word:        # if character is not in word this will run 
        chances-=1
        print("\nWrong Answer!")
        print(f"\nYou have {chances} chances left")
        if chances ==0:         # if there are no chances this condition will run
            print("\nYou Lose the game!!")
            break
```

```python
import random  
name=input("Enter you name: ")
print(f"\nHello {name}! welcome to guessing game!!")
chances=10
words=["apple","sunset","sunrise","calculator","bahubali","macbook","facebook","instagram","twitter","youtube","google","wipro","python"]
# getting a random word from the list
index=random.randint(0,len(words))
word=words[index]
indexes=random.sample(range(0,len(word)),3)
guesses=""   # character that the user guessed will stored here
for i in indexes:
    guesses+=word[i]

play="yes"

def playagain():    # 
    # here scoping concept
    global play
    play=input("Do you want to play again(yes/no) : ")
    if play=='yes':                # if your want to play agian this will exicute
        global chances,word,guesses  # here reseting all the new words and characters
        chances=10
        words=["apple","sunset","sunrise","calculator","bahubali","macbook","facebook","instagram","twitter","youtube","google","wipro","python"]
        index=random.randint(0,len(words))
        word=words[index]
        indexes=random.sample(range(0,len(word)),3)
        guesses=""
        for i in indexes:
            guesses+=word[i]     
    else:         # if user doesn't want to play this will print
        print("Thank you for playing")       
# this play is global variable    
while play=='yes':
    while chances>0:
        won=True
        for ch in word:
            if ch in guesses:
                print(ch,end=" ")
            else:
                print("_",end=" ")
                won=False
        if won:            # if you win this will run
            print("\nYou won the Game!")
            print(f"\nYour score is {chances*10}")
            playagain()  #calling function
            break
        guess=input("\nGuess a character: ")
        guesses+=guess
        if guess not in word:           # if character is not in word this will run 
            chances-=1
            print("\nWrong Answer")
            print(f"\nYou have {chances} chances left!")
            if chances==0:      # if there are no chances left this condition will run
                print("\nYou Lose the game!!")
                playagain()   # calling function 
                break
```
